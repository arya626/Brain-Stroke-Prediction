# Brain Stroke Prediction And Handling Imbalanced Data

## Introduction

The brain stroke is caused by the blockage of blood flow into the brain. There are a plenty of causes for stroke or heart failure including stress, heart disease, fat in blood, smoking etc. Although, we can classify brain stroke into five such as Ischemic, Hemorrhagic, Transient Ischemic Attack, Brain Stem Stroke, Cryptogenic Stroke, this project mainly focuses on predicting if the stroke occurs or not.

***This project is basically a binary classification problem where 1 indicates if the patient had a stroke, 0 otherwise.***
    
    

***Keywords***

__Supervised Learning  |  Binary Classification  |  Exploratory Data Analysis  |  Decision Tree  |  Random Forest Classifier  |  Support Vector Classifier  |  Naive Bayes  |  XGB Classifier  |  K Neighbors Classifier  |  Neural Network  |  Keras  |  Sequential Model  |  Synthetic Minority Oversampling Technique SMOTE__

## Data

The data for this project was obtained from Kaggle https://www.kaggle.com/datasets/fedesoriano/stroke-prediction-dataset <br>
The dataset is described as follows: <br>
This dataset originally has 4981 rows and 11 columns. <br>
The columns are described as follows:
<br>
1. gender: Male, Female or Other
2. age: Ranging from 8 to 82
3. hypertension: 1 Yes, 0 No
4. heart_disease: 1 Yes, 0 No
5. ever_married: 1 Yes, 0 No
6. work_type: Categorical values
7. Residence_type: Categorical values
8. avg_glucose_level: Continuous values
9. bmi: Body Mass Index (Continuous values)
10. smoking_status: 1 Yes, 0 No
11. stroke: Target variable (1 Yes, 0 No)
<br>

We will start with some analysis before moving onto the prediction and evaluation

##### We have few features that are of object types, it is necessary to convert them into type int before we begin building the model

```python
obj_cols = data.select_dtypes('object').columns
```

```python
obj_cols
```

```python
x = data.drop(['stroke'], axis = 1)
y = data['stroke']
```

##### Label Encoding is the best way to convert categorical values into numeric

```python
labelEn = LabelEncoder()
for col in obj_cols:
    x[col] = labelEn.fit_transform(x[col])
```

##### Splitting the data into train and test

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.25, random_state = 42)
```

##### Since I am experimenting with different classifiers, best practice is to define a function to train, test and evaluate the model

```python
def train_and_evaluate(model, xtrain, ytrain, xtest, ytest, algo):
    model.fit(xtrain, ytrain)
    predictions = model.predict(xtest)
    print("-"*20, algo,"-"*20)
    acc = np.round(accuracy_score(ytest, predictions),3)
    print("\nAccuracy: ", acc)
    print("\nClassification Report: \n")
    print(classification_report(ytest, predictions))
    print("\nConfusion Matrix: \n")
    cm = confusion_matrix(ytest, predictions, labels = model.classes_ )
    disp = ConfusionMatrixDisplay(confusion_matrix= cm, display_labels=model.classes_)
    disp.plot()
    plt.show()
    return acc
```

##### Defining the classifiers 

```python
modeldt = DecisionTreeClassifier()
modelrf = RandomForestClassifier()
modelnb = GaussianNB()
modelsv = SVC()
modelxgb = XGBClassifier()
modelkn = KNeighborsClassifier()
models = [(modeldt, "Decision Tree Classifier"), 
          (modelrf, "Random Forest Classifier"),
          (modelnb, "Naive Bayes"),
          (modelsv, "Support Vector Classifier"),
          (modelxgb, "XGBoost"),
         (modelkn, "KNN Classifier")]
all_acc = {}
for model in models:
    acc = train_and_evaluate(model[0], xtrain, ytrain, xtest, ytest, model[1])
    all_acc[model[1]] = acc
```

```python
plt.figure(figsize = (15,8))
plt.bar(all_acc.keys(),all_acc.values())
```

##### From the above plot we can blindly conclude that the best model is Support Vector Classifier as it provides an accuracy of 94.5%. However, accuracy isn't enough to evaluate our model. Hence we look into the classification report to judge how well our model is performing.

##### From the Classification Reports we can see that Precision, Recall and F1-Score for the class 1 is very small as compared to the class 0 
##### This is mainly because of model overfitting. It is caused by the imbalanced data
##### For any model to perform at its best, it is important for the data to be balanced

##### Let us check how imbalanced our data is

```python
data['stroke'].value_counts()
```

```python
plt.figure(figsize = (12,8))
sns.countplot(data['stroke'])
```

##### There are 4733 records that belongs to class 0 and 248 belongs to class 1. This causes the model to be biased towards class 0. 
##### The model is performing with high accuracy because the test data has tiny number of class 1 and the model is predicting all inputs in the test data as 0
##### One way to solve this issue is by oversampling the data. This means creating new samples that belongs to the minority class
##### Our goal of this project is to provide more number of true positives & True Negatives and reduce the number of False Positives & False Negatives. Our focus should lie on reducing the number of False Negatives as determining storke as no-stroke could lead to high risk

### Oversampling

```python
over_sample = SMOTE()
```

```python
X,Y = over_sample.fit_resample(x,y)
```

```python
X['age_rounded'] = np.round(X['age'])
```

```python
X.drop(['age'], inplace = True, axis = 1)
```

```python
X
```

```python
xtrain1, xtest1, ytrain1, ytest1 = train_test_split(X,Y, test_size = 0.25, random_state = 42)
```

```python
all_acc1 = {}
for model in models:
    acc = train_and_evaluate(model[0], xtrain1, ytrain1, xtest1, ytest1, model[1])
    all_acc1[model[1]] = acc
```

```python
model1 = Sequential()
model1.add(Dense(100, input_dim = 10 , activation='relu'))
model1.add(Dense(100, activation = 'relu'))
# model1.add(Dense(60, activation = 'sigmoid'))
model1.add(Dense(50, activation = 'relu'))
model1.add(Dense(50, activation = 'relu'))
model1.add(Dense(30, activation= 'relu'))
model1.add(Dense(20, activation = 'relu'))
model1.add(Dense(1, activation = 'sigmoid'))

model1.summary()
model1.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='Adam')
```

```python
# es = EarlyStopping(monitor='loss', mode='min')
hist = model1.fit(xtrain1,ytrain1, epochs=500, validation_data = (xtest1, ytest1))
```

```python
_, train_acc = model1.evaluate(xtrain1, ytrain1, verbose=0)
_, test_acc = model1.evaluate(xtest1, ytest1, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
```

```python
plt.figure(figsize=(15,8))
plt.plot(hist.history['accuracy'],label = "Train")
plt.plot(hist.history['val_accuracy'], label = "Validation")
plt.legend(loc = "best")
plt.show()
```
